```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= np.linspace(0,2*np.pi,100)
cosdata=np.cos(data)

plt.plot(data,cosdata,marker="o",color="blue")
```

```python
def embed(points,window=2):
    n=len(points)
    X=np.zeros((n-window,window))
    for i in range(n-window):
        X[i]=points[i:i+window]
    return X
```

```python
embedded = embed(cosdata,window=2)
plt.plot(embedded[:,0],embedded[:,1],marker="o",color="red")
```

```python
sindata = np.sin(data)
embedded = embed(sindata,window=2)
plt.plot(embedded[:,0],embedded[:,1],marker="o",color="green")
```

```python
linedata = 3*data+2
embeddedline= embed(linedata,window=2)
plt.plot(embeddedline[:,0],embeddedline[:,1],marker="o",color="black")
```
